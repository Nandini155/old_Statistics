```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# make data
# from statistics import mode

np.random.seed(42) # For reproducibility , ensures consistent results across multiple runs.
data=np.random.beta(1,3,10000) # Generate data with Beta(1, 3)
raw=pd.DataFrame(data,columns=["value"])

print("mean :",raw['value'].mean())
print("median :",raw['value'].median())
print("standard deviation:", raw['value'].std())
print("variance:", raw['value'].var())
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(data=raw['value'], bins=30, kde=True)
plt.title("Original Beta(1, 3) Distribution")
plt.show()
```

```python
# shift the data to avoid zeros or values close to zero
# Shift data to avoid zeros (Box-Cox limitation)
data_shifted = data + 1e-8  # Shift to ensure positive values
```

```python
from scipy import stats
fitted_data, fitted_lambda = stats.boxcox(data_shifted)
print(f"Box-Cox Lambda: {fitted_lambda}")
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(data=fitted_data,bins=30,kde=True)
plt.title("Boxcox Transformation")
plt.show()
```
